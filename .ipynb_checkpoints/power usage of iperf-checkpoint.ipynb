```python
import numpy as np
import pandas as pd
import plorts
import matplotlib.pyplot as plt
plt.style.use(['plorts', 'plorts-web'])
```

```python
df_1m = pd.read_csv('power-log-1m-combined.csv', sep=' ', names=['intel-rapl:0', 'intel-rapl:1'])
df_full = pd.read_csv('power-log-full-combined.csv', sep=' ', names=['intel-rapl:0', 'intel-rapl:1'])
df_nothing = pd.read_csv('power-log-nothing-combined.csv', sep=' ', names=['intel-rapl:0', 'intel-rapl:1'])
```

```python
df_1m['time'] = df_1m.index
df_1m['intel-rapl:0'] -= df_1m['intel-rapl:0'].min()
df_1m['intel-rapl:1'] -= df_1m['intel-rapl:1'].min()
```

```python
df_full['time'] = df_full.index
df_full['intel-rapl:0'] -= df_full['intel-rapl:0'].min()
df_full['intel-rapl:1'] -= df_full['intel-rapl:1'].min()
```

```python
df_nothing['time'] = df_nothing.index
df_nothing['intel-rapl:0'] -= df_nothing['intel-rapl:0'].min()
df_nothing['intel-rapl:1'] -= df_nothing['intel-rapl:1'].min()
```

```python
plt.plot(df_1m['time'], df_1m['intel-rapl:1'], label='1m')
plt.plot(df_full['time'], df_full['intel-rapl:1'], label='full')
plt.plot(df_nothing['time'], df_nothing['intel-rapl:1'], label='nothing')
plt.legend(loc='best')
```

```python
1.0*(df_1m.loc[300]['intel-rapl:1'] - df_full.loc[300]['intel-rapl:1'])/df_full.loc[300]['intel-rapl:1']
```

```python
plt.plot(df_1m['time'], df_1m['intel-rapl:0'], label='1m')
plt.plot(df_full['time'], df_full['intel-rapl:0'], label='full')
plt.plot(df_nothing['time'], df_nothing['intel-rapl:1'], label='nothing')
plt.legend(loc='best')
```
